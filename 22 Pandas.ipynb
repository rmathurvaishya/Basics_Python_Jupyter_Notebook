###  Merge Dataframes

```python
import pandas as pd

# MODEL 1

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
# MODEL 2

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how ='outer',indicator=True)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right",indicator=True) #d2 ke all & d1 se common
df3
```

#### How to use "suffixes()" argument in dataframes?

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
d3 = pd.merge(df1,df2,on='city')
d3
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```

### RESAMPLE

##### This process of changing the time period that data are summarized for is often called resampling.
##### Object must have a datetime-like index. 
##### Two types of resampling are:
        Upsampling: Where you increase the frequency of the samples, such as from minutes to seconds.
        Downsampling: Where you decrease the frequency of the samples, such as from days to months.

```python
apple = pd.read_csv("E:/New folder/AnalytixLabs/Python/case study/appl_1980_2014.csv")
apple.head()
```

```python
apple.Date = pd.to_datetime(apple.Date)
```

```python
apple = apple.set_index('Date')

apple.head()
```

```python
## Get the last business day of each month

last_day = apple.resample('BM').mean()

last_day.head()
```

```python
last_day = apple.resample('BM').mean()

len(last_day)
```
